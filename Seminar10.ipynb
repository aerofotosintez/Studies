```python
corr = penguins.corr()
sb.heatmap(corr,cmap='magma')
```

```python
corr = penguins.corr()
mask = np.triu(np.ones_like(corr,dtype=bool))
#создаем полотно для отображения большого графика
f,ax=plt.subplots(figsize=(11,9))
#создаем цветовую политру
cmap=sb.diverging_palette(230,20,as_cmap=True)
#визуализируем данные кореляции
sb.heatmap(corr,mask=mask,cmap=cmap,vmax=.3,center=0,square=True,linewidths=.5,cbar_kws={'shrink':.5})
```

пингвинами, который будет отвечать за
показатель длины клюва пингвина.

middle - средний(от 35 до 42)
low - маленький(до 35)

```python
penguins = sb.load_dataset("penguins")
penguins.head()
```
